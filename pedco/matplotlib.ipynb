```python
import matplotlib.pyplot as plt


plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.ylabel('probando números')
plt.xlabel('probando otros números')
plt.show()
```

```python
# con la cadena de formato que indica el color y el tipo de línea del gráfico. 
#Las letras y símbolos de la cadena de formato son de MATLAB, y se concatena una cadena de color con una cadena de estilo de línea. 
#La cadena de formato predeterminada es 'b-', que es una línea azul sólida. 
# Por ejemplo, para trazar lo anterior con círculos verdes:

plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'go-', linewidth=1)
plt.ylabel('probando números')
plt.xlabel('probando otros números')
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
#xmin, xmax, ymin, ymax = axis([xmin, xmax, ymin, ymax])
plt.axis([0, 6, 0, 30])
plt.show()
```

```python
import numpy as np

# definir un arreglo en nympy
t = np.arange(0., 5., 0.2)
print(t)

# dibujar el mismo arreglo en formato distinto y multilpicando cada valor por 1, 2 o 3 respectivamente
plt.plot(t, t*1, 'r--', t, t*2, 'bs', t, t*3, 'g^')
plt.show()
```

```python
# Ploteando con variable categóricas varias subfiguras
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(figsize=(9, 3))

plt.subplot(131)
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values, marker='^')
plt.subplot(133)
plt.plot(names, values,'g')
plt.suptitle('Categorical Plotting')
plt.show()
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]
```

```python
#agregando texto
ax = plt.subplot()

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
line, = plt.plot(t, s, lw=2)

plt.annotate('aca quiero que vean', xy=(2, 1), xytext=(3, 1.5),
             arrowprops=dict(facecolor='black', shrink=0.05),
             )

plt.annotate('y de paso aca tmb', xy=(3.5, 0.0), xytext=(3, -1.5),
             arrowprops=dict(facecolor='green', shrink=0.05),
             )

plt.ylim(-2, 2)
plt.show()
```
